```python
# Existe una diferencia entre los parametros y el soporte
# CONCEPTO DE FUNCION

def posicion(t,x0,v):
  # la fisica
  x = x0 + v*t
  return x
```

```python
posicion(0,1.2,7)
```

```python
# Vamos a generar el soporte
t = []
x = []
for i in range (10):
  t.append(i)
  x.append(posicion(i,1.2,7))
```

```python
# Ejercicio en donde se cruzan dos carritos
t = []
x = []
x1 = []
for i in range (10):
  t.append(i)
  x.append(posicion(i,1.2,7))
  x1.append(posicion(i,20,-3))
```

```python
plt.plot(t,x)
plt.plot(t,x1)
plt.show()
```
