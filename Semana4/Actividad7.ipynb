```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
import time
import copy
```

```python
f = lambda x,y,z : (x-1)**2 + 2*(y+2)**2 + 3*(z-1)**2

#Cálculo numérico de las derivadas parciales de cada variable

def PartialDerivativeX(x,y,z,f,h=1e-4):
    return (f(x+h, y, z) - f(x-h, y, z)) / (2*h)

def PartialDerivativeY(x,y,z,f,h=1e-4):
    return (f(x, y+h, z) - f(x, y-h, z)) / (2*h)

def PartialDerivativeZ(x,y,z,f,h=1e-4):
    return (f(x, y, z+h) - f(x, y, z-h)) / (2*h)

xvalues = np.linspace(-10,10,50)
yvalues = np.linspace(-10,10,50)
zvalues = np.linspace(-10,10,50)

def GradientDescent(f, epochs=50,lr=0.1):

  vn = np.array([np.random.uniform(-10,10), np.random.uniform(-10,10), np.random.uniform(-10,10)]) #Se tomaron semillas aleatorias para las 3 variables
  History = np.array([])

  for e in range(epochs):

    Grad = np.array([PartialDerivativeX(vn[0],vn[1],vn[2],f),
                     PartialDerivativeY(vn[0],vn[1],vn[2],f),
                     PartialDerivativeZ(vn[0],vn[1],vn[2],f)])

    vn1 = vn - lr*Grad
    vn = vn1

    History = np.append(History, f(vn[0], vn[1], vn[2]))

    clear_output(wait=True)

    plt.plot(History,'.-',color='k')
    plt.show()

  return vn

GradientDescent(f)

"""El mínimo hallado es x = 0.999, y = -2, z = 1, se aproxima bastante al valor esperado para que la función llegue a cero"""
```
