```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
import time
import copy
```

```python
def Plotter(e):

  fig = plt.figure(figsize=(8,3))
  ax = fig.add_subplot(121)
  ax1 = fig.add_subplot(122)
  ax.set_title('iteracion: {:.0f}'.format(e),fontsize=10)

  return ax,ax1
```

```python
ax,ax1 = Plotter(6)
```

```python
def f(x,y,z):
  return ((x-1)**2)+ (2(y+2)**2) + (3(z-1)**2)
```

```python
def DerivativeX(x,y,z,f,h=1e-4):
  return (f(x+h,y,z)-f(x-h,y,z))/(2*h)
```

```python
def DerivativeY(x,y,z,f,h=1e-4):
  return (f(x,y+h,z)-f(x,y-h,z))/(2*h)
```

```python
def DerivativeZ(x,y,z,f,h=1e-4):
  return (f(x,y,z+h)-f(x,y,z-h))/(2*h)
```

```python
Grad = np.zeros((1,3))
Grad
```

```python
historia = np.array([])

xn = 0.5
yn = 2
zn = 1.5
lr = 0.2

for e in range(100):
  Grad[0,0] = DerivativeX(xn,yn,zn)
  Grad[0,1] = DerivativeY(xn,yn,zn)
  Grad[0,2] = DerivativeZ(xn,yn,zn)

  xn1 = xn - lr*Grad
  yn1 = yn - lr*Grad
  zn1 = zn - lr*Grad

  historia = np.append(historia,f(xn1,yn1,zn1))

  xn = xn1
  yn = yn1
  zn = zn1

print(xn,yn,zn)


```
