```python
import numpy as np
import matplotlib.pyplot as plt

def funcion(x,k1,k2):
  y = abs(x)**(2/3) + k1*np.sin(10*k2*x)*np.sqrt(3-x**2)
  return y
```

```python
x = np.linspace(-np.sqrt(3), np.sqrt(3), 1000)
y = funcion(x,1.2,4.4)

plt.plot(x,y, color='red')
plt.show()
```
