```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Funcion anonima
s = lambda n: 1/n
```

```python
s(2)
```

```python
# Quiero evaluar la sucesion en varios puntos
N = 20
n = np.arange(1, N+1)
n
# Dominimo (eje x)
```

```python
a = np.zeros(N)
a
```

```python
for i in range(n.shape[0]):
  a[i] = s(i+1)
```

```python
a
```

```python
plt.plot(n, a, marker='o')
```

```python
# Problema computacional: guardar en una lista la distancia entre puntos. Me interesa la distancia vertical
np.diff(a)
# Es negativa porque va bajando la sucesion
```

```python
# Verificacion de que es sucesion de cauchy
# Otra forma de plotear datos
plt.stem(np.diff(a))
```

```python
# Quiero saber el N para el que la vecindad (e) es  e < 1x10**-3
epsilon = 1e-3

diff = 1000
n = 1

while diff > epsilon:
  vecinoizq = s(n)
  vecinoder = s (n+1)
  diff = np.abs(vecinoizq-vecinoder)
  n += 1
  print(n, diff)

  # Condicion de parada
  if diff < epsilon:
    print("terminamos: ", n)
    break
```
