```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
s = lambda n: (-1)**n
```

```python
N = 50
n = np.arange(1, N+1)

c = np.zeros(N)

for i in range(n.shape[0]):
  c[i] = s(i+1)
```

```python
c
```

```python
plt.plot(n, c, marker = "o")
```

```python
z = lambda n: 1/2**n
```

```python
N = 50
p = np.arange(1, N+1)

d = np.zeros(N)

for i in range(1, p.shape[0]):
  for j in range(1, i+1):
    d[i] += z(j)
```

```python
d
```

```python
plt.plot(p, d, marker = "o")
```
