```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
N = 20
x = np.linspace(-5, 5, N)
y = np.linspace(-5, 5, N)
```

```python
X,Y = np.meshgrid(x, y)
```

```python
Ex = np.zeros((N,N))
Ey = np.zeros((N,N))
```

```python
def GetField(x,y,r0):

  x0, y0 = r0

  # Ley de Coulomb
  Ex = (x-x0)/( (x-x0)**2+ (y-y0)**2 )**(3/2)
  Ey = (y-y0)/( (x-x0)**2+ (y-y0)**2 )**(3/2)

  return Ex, Ey
```

```python
# Crear la estructura de la fuente
fuentes = np.zeros((4,2))
fuentes
```

```python
fuentes[0] = [0,2]
fuentes[1] = [-1,-4]
fuentes[2] = [-4,4]
fuentes[3] = [2,-2]
fuentes
```

```python
# Llenamos el valor del campo
for k in range(fuentes.shape[0]):
  for i in range(N):
    for j in range(N):
      Ex[i,j] += GetField(x[i],y[j],fuentes[k])[0]
      Ey[i,j] += GetField(x[i],y[j],fuentes[k])[1]
```

```python
plt.scatter(X,Y)
plt.scatter(*fuentes[0], color="r", s=200)
plt.scatter(*fuentes[1], color="r", s=200)
plt.scatter(*fuentes[2], color="r", s=200)
plt.scatter(*fuentes[3], color="r", s=200)

for i in range(N):
  for j in range(N):
    plt.quiver(x[i],y[j],Ex[i,j],Ey[i,j])
```
