```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Necesitamos imprimir el conjunto de soporte del campo
# Vamos a definir UN LATTICE
N = 20
x = np.linspace(-5, 5, N)
y = np.linspace(-5, 5, N)
```

```python
x
```

```python
y
```

```python
X,Y = np.meshgrid(x, y)
```

```python
plt.scatter(X,Y)
```

```python
# Vamos a pintar un campo escalar
def f2(x, y):
  return np.sin(x) + 5*np.cos(y)
```

```python
# Definimos la imagen
Z = np.zeros_like(X)
```

```python
Z
```

```python
# Es el valor del campo escalar sobre todo el lattice
for i in range(N):
  for j in range(N):
    Z[i,j] = f2(x[i], y[j])
```

```python
Z
```

```python
from matplotlib import cm
fig1 = plt.figure(figsize=(3,3))
ax1 = fig1.add_subplot(111, projection='3d')
ax1.plot_surface(X, Y, Z, cmap="viridis")
```

```python
# Para un campo vectorial necesitamos 2 componentes
Ex = np.zeros((N,N))
Ey = np.zeros((N,N))
```

```python
Ey
```

```python
# Vamos a definir que es el campo
def GetField(x,y,r0):

  x0, y0 = r0

  # Ley de Coulomb
  Ex = (x-x0)/( (x-x0)**2+ (y-y0)**2 )**(3/2)
  Ey = (y-y0)/( (x-x0)**2+ (y-y0)**2 )**(3/2)

  return Ex, Ey
```

```python
fuentes = [0.1,2]
```

```python
GetField(1,3,fuentes)
```

```python
# Llenamos el valor del campo
for i in range(N):
  for j in range(N):
    Ex[i,j] = GetField(x[i],y[j],fuentes)[0]
    Ey[i,j] = GetField(x[i],y[j],fuentes)[1]
```

```python
plt.scatter(X,Y)
plt.scatter(*fuentes, color="r", s=200)

for i in range(N):
  for j in range(N):
    plt.quiver(x[i],y[j],Ex[i,j],Ey[i,j])
```
