```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
import sympy as sym
```

```python
# Necesitamos imprimir el conjunto de soporte del campo
# Vamos a definir UN LATTICE
N = 100
x = np.linspace(-1, 1, N)
y = np.linspace(-1, 1, N)
```

```python
X,Y = np.meshgrid(x, y)
```

```python
# Vamos a pintar un campo escalar
def f2(x, y):
  return np.sqrt(1- x**2 - y**2)
```

```python
# Definimos la imagen
Z = np.zeros_like(X)
```

```python
# Es el valor del campo escalar sobre todo el lattice
for i in range(N):
  for j in range(N):
    Z[i,j] = f2(x[i], y[j])
```

```python
from matplotlib import cm
fig1 = plt.figure(figsize=(5,5))
ax1 = fig1.add_subplot(111, projection='3d')
ax1.plot_surface(X, Y, Z, cmap="viridis")
```
