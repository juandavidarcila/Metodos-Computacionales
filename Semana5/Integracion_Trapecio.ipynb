```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
import sympy as sym
```

```python
x_ = sym.Symbol('x',real=True)
F = sym.sin(2*x_)*sym.exp(-x_)
F
```

```python
sym.integrate(F,(x_,0,2*np.pi))
```

```python
a = 0*-2*np.pi
b = 2*np.pi
f = lambda x: np.sin(2*x)*np.exp(-x)
x = np.linspace(a,b,30)
y = f(x)
```

```python
N = 8 # Controlamos el numero de trapecios
x1 = np.linspace(a,b, N+1)
y1 = f(x1)
yp = np.interp(x1,x1,y1)
```

```python
plt.plot(x,y,color='k')
plt.plot(x1,yp,marker='*',color='r')
#plt.fill_between(x,y,color='b',alpha=0.5)
plt.fill_between(x1,yp,color='b',alpha=0.5)
```

```python
# Calculamos h
h = x1[1] - x1[0]
h
```

```python
x1
```

```python
Integral = 0
Integral += 0.5*(f(x1[0]) + f(x1[-1]))

for i in range (1,len(x1)-1):
    Integral += f(x1[i])

Integral *= h

Integral
```
