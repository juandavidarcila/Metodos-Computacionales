```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Function(x):
  return x**x - 100
```

```python
# Primer paso: pintamos la funcion
x = np.linspace(1, 5, 50)
y = Function(x)
```

```python
plt.plot(x, y)
plt.axhline(y=0, color='r', linestyle='--')
```

```python
# Como un curso de IP
for i in range(len(x)-1):
  print(x[i])
```
