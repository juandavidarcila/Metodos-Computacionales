```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-np.pi,np.pi,20)
x
```

```python
def funcion(x):
  return np.sin(x)

y = funcion(x)
plt.scatter(x,y)
```

```python
# Queremos definir la derivada derecha
def RightDerivate(x,f,h=0.2):

  d = (f(x+h)-f(x))/h

  return d
```

```python
# Queremos definir la derivada izquierda
def LeftDerivate(x,f,h=0.2):

  d = (f(x)-f(x-h))/h

  return d
```

```python
# Queremos definir la derivada central
def CentralDerivate(x,f,h=0.2):

  d = (f(x+h)-f(x-h))/(2*h)

  return d
```

```python
LeftDerivate(np.pi/2,funcion)
```

```python
derivada1 = RightDerivate(x,funcion)
derivada2 = LeftDerivate(x,funcion)
derivada3 = CentralDerivate(x,funcion)
plt.plot(x,derivada1)
plt.plot(x,derivada2)
plt.plot(x,derivada3)
plt.plot(x,np.cos(x), label = "Exacta")
plt.legend()
```

```python
# Quiero una derivada parcial
def f2(x,y):
  return x**2 + y**2
```

```python
def PartialDerivate(x,y,f2,h=0.001):

  d = (f2(x+h,y)-f2(x-h,y))/(2*h)

  return d
```

```python
PartialDerivate(1,2,f2)
```
