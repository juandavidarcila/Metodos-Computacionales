```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-np.pi,np.pi,20)
x
```

```python
def funcion(x):
  return np.sin(x)

y = funcion(x)
plt.scatter(x,y)
```

```python
# Queremos definir la derivada derecha
def RightDerivate(x,f,h=0.2):

  d = (f(x+h)-f(x))/h

  return d
```

```python
RightDerivate(np.pi/2,funcion)
```

```python
derivada1 = RightDerivate(x,funcion)
plt.plot(x,derivada1)
plt.plot(x,np.cos(x), label = "Exacta")
plt.legend()
```
