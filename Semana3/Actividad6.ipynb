```python
import numpy as np
import matplotlib.pyplot as plt

"""a) Después de ver la gráfica, podría decir que el intervalo en donde ocurre el encuentro es [5, 6]"""

def x1(t):
  return 50 - 0.2*t + 0.1*t**2

def x2(t):
  return t**2 + 0.1*t**3

t = np.linspace(0, 7, 100)
plt.plot(t, x1(t))
plt.plot(t, x2(t))

"""Defino f(t) como errorFun, esta es la función que queremos minimizar/ hallar cero. Busca el punto en el que las funciones son iguales, que al restarse dan cero


"""

def errorFun(t):
  return x1(t) - x2(t)

def Derivative(x,f,h=0.1e-6):
  d = (f(x+h) - f(x-h))/(2*h)
  return d

def GetNewtonMethod(f,df,xn,itmax=100,precision=1e-10):

  error = 1
  it = 0

  while error > precision and it < itmax:

    try:
      xn1 = xn - f(xn)/df(xn,f)
      print(xn,xn1)
      error = np.abs(xn1 - xn) # La nocion de cauchy
    except ZeroDivisionError:
      print("Division por cero")

    xn = xn1
    it += 1

  if it == itmax:
    return False
  else:
    return xn

"""Como semilla escogi el límite inferior del rango que seleccioné anteriormente, como criterio de parada coloqué una presición de 1e-10"""

GetNewtonMethod(errorFun, Derivative, 5.)

t_prima = 5.754043704211815

plt.scatter(t_prima, x2(t_prima), color = 'red')
plt.plot(t, x1(t)); plt.plot(t, x2(t))

display(x1(t_prima))
display(x2(t_prima))

x1(t_prima) == x2(t_prima) #A pesar de la precision no es completamente igual jeje
```
