```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
N = 25
x = np.linspace(-4, 4, N)
y = np.linspace(-4, 4 , N)
```

```python
def FuncionFlujo(x,y,R=2, v=2):
  return v*x*(1-((R**2)/(x**2)+y**2))
```

```python
def PartialDerivate(x,y,FuncionFlujo,h=0.001):

  vx = (FuncionFlujo(x+h,y)-FuncionFlujo(x-h,y))/(2*h)
  vy = (FuncionFlujo(y+h,x)-FuncionFlujo(y-h,x))/(2*h)

  return vx, vy
```

```python
X,Y = np.meshgrid(x, y)
```

```python
Vx = np.zeros((N,N))
Vy = np.zeros((N,N))
```

```python
for i in range(N):
  for j in range(N):
    Vx[i,j], Vy[i,j] = PartialDerivate(X[i,j],Y[i,j],FuncionFlujo)
```

```python
plt.scatter(X,Y)

for i in range(N):
  for j in range(N):
    plt.quiver(x[i], y[j], Vx[i,j], Vy[i,j], color="blue")
```
